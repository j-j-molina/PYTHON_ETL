# Ciencia de Datos en Producción — Entregable 3  
## Cargar_datos.ipynb — Materialización del dataset (XLSX → CSV)

> **Recibe el archivo “puro” en XLSX** y **lo materializa como CSV** para que el resto del pipeline lo consuma.

**Entrada:** `BD_creditos (2).xlsx` (archivo crudo)  
**Salida:** `Base_de_datos.csv` (dataset completo en CSV, guardado en `mlops_pipeline/src/`)  

---

```python
import pandas as pd
from pathlib import Path

pd.set_option("display.max_columns", 200)
pd.set_option("display.width", 140)


def find_repo_root(start: Path) -> Path:
    """Intenta encontrar la raíz del repo (donde exista la carpeta mlops_pipeline)."""
    p = start.resolve()
    for _ in range(8):
        if (p / "mlops_pipeline").exists():
            return p
        p = p.parent
    return start.resolve()
```

```python
# 1) Ubicar el XLSX crudo
here = Path.cwd()
repo_root = find_repo_root(here)

candidates = [
    here / "BD_creditos (2).xlsx",
    repo_root / "BD_creditos (2).xlsx",
    repo_root / "mlops_pipeline" / "src" / "BD_creditos (2).xlsx",
    repo_root / "src" / "BD_creditos (2).xlsx",
]

xlsx_path = next((p for p in candidates if p.exists()), None)
if xlsx_path is None:
    raise FileNotFoundError(
        "No encontré 'BD_creditos (2).xlsx'.\n"
        "Ubica el archivo en: mlops_pipeline/src/BD_creditos (2).xlsx\n"
        "ó actualiza la variable 'xlsx_path' con la ruta correcta."
    )

print("✅ XLSX encontrado en:", xlsx_path)
```

```python
# 2) Cargar XLSX
df_raw = pd.read_excel(xlsx_path)

print("✅ Dataset cargado")
print("Shape:", df_raw.shape)
df_raw.head()
```

```python
# 3) Exportar a CSV (dataset completo)
out_dir = repo_root / "mlops_pipeline" / "src"
out_dir.mkdir(parents=True, exist_ok=True)
csv_path = out_dir / "Base_de_datos.csv"

df_raw.to_csv(csv_path, index=False, encoding="utf-8")

print("✅ CSV guardado en:", csv_path)
print("Columnas:", len(df_raw.columns))
```
